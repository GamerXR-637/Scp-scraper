```python
import requests
from bs4 import BeautifulSoup
import re
import json

# Define the harmonize_id function
def harmonize_id(id):
    return f"{id:03d}"  # Converts the ID to a zero-padded 3-digit string

# Define the relax_key function
def relax_key(key):
    return key.lower().replace(' ', '_')  # Example: Converts "Object Class" to "object_class"

# Define the affix_additional function
def affix_additional(data):
    # Example: Add additional metadata or processing to the result dictionary
    data['processed'] = True
    return data

# Define the base_url variable
base_url = "http://www.scpwiki.com/scp-"
from bs4 import BeautifulSoup
import re
import json

# Ask the user to input the ending ID
end = int(input("Enter the ID where the scraper should stop: "))

# Main scraping function
def scrape_scp(id):
    result_dict = {}
    _id = harmonize_id(id)
    url = base_url + _id
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx and 5xx)
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return {}
    print(f"Fetching: {url}")
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the page content
    page_content = soup.find('div', id='page-content')
    if not page_content:
        print(f"Error: Could not find page content for SCP-{_id}")
        return {}

    paragraphs = page_content.find_all('p')

    # Process or extract the desired data from the paragraphs
    current_key = None
    current_value = ''
    stop = False

    for paragraph in paragraphs:
        if not stop:
            strong_tag = paragraph.find('strong')

            if strong_tag:
                if current_key:
                    _k = relax_key(current_key)
                    current_value = re.sub(r'\u2588+', '[REDACTED]', current_value)
                    
                    # Log the description field
                    if _k == 'description':
                        print(f"SCP-{_id} Description: {current_value.strip()}")

                    result_dict[_k] = current_value.strip()

                current_key = strong_tag.get_text()
                current_value = paragraph.get_text()[len(current_key):].strip()

            elif paragraph.get_text().startswith('Â«'):
                stop = True

            else:
                if current_value:
                    current_value += ' '
                current_value += paragraph.get_text()

    # Add the last key-value pair
    if current_key:
        _k = relax_key(current_key)
        current_value = re.sub(r'\u2588+', '[REDACTED]', current_value)
        
        # Log the description field
        if _k == 'description':
            print(f"SCP-{_id} Description: {current_value.strip()}")

        result_dict[_k] = current_value.strip()

    # Extract the SCP name (if available)
    title_tag = soup.find('title')
    if title_tag:
        result_dict['name'] = title_tag.get_text().split(' - ')[0]  # Extract the name before " - SCP Foundation"

    result_dict = affix_additional(result_dict)
    return result_dict

# Mass scraping block
file_path = "scp_database.json"

with open(file_path, "w", encoding="utf-8") as file:
    for i in range(1, end + 1):
        scp_data = scrape_scp(i)
        if scp_data:  # Only write if data is not empty
            file.write(json.dumps({str(i): scp_data}, indent=4) + "\n")

print(f"Dictionary written to {file_path} successfully.")
# Define the harmonize_id function
def harmonize_id(id):
    return f"{id:03d}"  # Converts the ID to a zero-padded 3-digit string

# Define the relax_key function
def relax_key(key):
    return key.lower().replace(' ', '_')  # Example: Converts "Object Class" to "object_class"

# Define the affix_additional function
def affix_additional(data):
    # Example: Add additional metadata or processing to the result dictionary
    data['processed'] = True
    return data

# Define the base_url variable
base_url = "http://www.scpwiki.com/scp-"

# Import necessary modules
```
